```python
text = "Hello world 123"
n = 2  # n-gram size

# Generate word-level n-grams
words = text.split()
word_ngrams = list(ngrams(words, len(words)))

print(word_ngrams)  # output: [('Hello', 'world')]
```

```python
def split_location_into_ngrams(location):
    words = location.split(' ')
    if len(words) > 2:
        return [' '.join(x) for x in zip(words, words[1:])]
    return None
```

```python
sal2 = sal.copy()
sal2['loc'] = sal2.location.apply(lambda x: split_location_into_ngrams(x))
sal2
```
